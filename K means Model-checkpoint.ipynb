```python
# Importing necessary libraries
import pandas as pd  # For data analysis
import seaborn as sns  # For statistical visualization
import matplotlib.pyplot as plt  # For basic visualization
import plotly.express as px  # For interactive visualization
import numpy as np  # For numerical computing
```

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t") # importing the data set in the form of data frame and seperated with tab space 
df.head() # will print the first 5 from top values or data from teh data set  
```

```python
df["TotalAmountSpent"] = df["MntFishProducts"] + df["MntFruits"] + df["MntGoldProds"] + df["MntSweetProducts"] + df["MntMeatProducts"] + df["MntWines"]
# this will create new column 'TotalAmountSpent' which will contain total amount spent by the user 
df.head() # very last column is 'TotalAmountSpent'
```

```python
# Filling missing values in the "Income" column with the median value
df["Income"].fillna(df["Income"].median(), inplace=True)

# Selecting columns "Income" and "TotalAmountSpent"
data = df[["Income", "TotalAmountSpent"]]

# Applying logarithmic transformation to the selected data
df_log = np.log(data)
```

```python
# StandardScaler is used to scale the features to have a mean of 0 and a variance of 1.
# This is important for many machine learning algorithms that assume data is centered and has a consistent scale.
from sklearn.preprocessing import StandardScaler

# KMeans is a clustering algorithm used to partition data into clusters based on similarity.
# It aims to group data points such that points in the same group (cluster) are more similar to each other than to those in other groups.
from sklearn.cluster import KMeans
```

```python
# Initialize a StandardScaler object
std_scaler = StandardScaler()

# Scale the log-transformed data
df_scaled = std_scaler.fit_transform(df_log)
```

```python
errors = []
for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(df_scaled)
    errors.append(model.inertia_)

```

```python
# Set the title of the plot
plt.title('The Elbow Method')

# Set the labels for the x-axis and y-axis
plt.xlabel('k')
plt.ylabel('SSE')  # SSE: Sum of Squared Errors

# Create a point plot using Seaborn, with k values ranging from 1 to 10
# and corresponding SSE values stored in the errors list
sns.pointplot(x=list(range(1, 11)), y=errors)

# Save the plot as an image file named "K_Mean_Elbow.png"
plt.savefig("K_Mean_Elbow.png")
```

```python
from kneed import KneeLocator

# Create a KneeLocator object to identify the elbow point in the curve
kl = KneeLocator(x=range(1, 11),
                 y=errors,
                 curve="convex",  # Specifies the shape of the curve
                 direction="decreasing")  # Specifies the direction of the curve

# Print the optimum number of clusters identified by the KneeLocator
print('The optimum number of clusters is: ' + str(kl.elbow + 1 ))
```

```python
model = KMeans(n_clusters=3, random_state=42)  # Create a KMeans clustering model with 3 clusters
model.fit(df_scaled)  # Fit the model to the scaled data
```

```python
data = data.assign(ClusterLabel=model.labels_)  # Assign cluster labels to the data
cluster_summary = data.groupby("ClusterLabel")[["Income", "TotalAmountSpent"]].median()  # Group data by cluster label and calculate median income and total amount spent
cluster_summary
```

```python
# Create a scatter plot using Plotly Express
fig = px.scatter(
    data_frame=data,  # DataFrame containing the data
    x="Income",       # X-axis variable
    y="TotalAmountSpent",  # Y-axis variable
    title="Relationship between Income VS Total Amount Spent",  # Title of the plot
    color="ClusterLabel",  # Color the points based on cluster labels
    height=500       # Height of the plot
)
fig.show()  # Display the plot
```

## Interpreting the cluster Label.

- Cluster 0: Customers with moderate Income and Moderate spending.
- Cluster 1: Customers with low Income and Low spending.
- Cluster 2: Customers who earn much and spend much.
